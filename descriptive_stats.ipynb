```python
import pandas as pd
import numpy as np
from os import getcwd
%load_ext autoreload
%autoreload 2

# Load the data
df = pd.read_csv("filtered.csv")
```

```python
# Balance Sheet Summary Statistics
df_shareholders = df[{'Assets', 'LiabilitiesAndStockholdersEquity', 'RetainedEarningsAccumulatedDeficit', 'CashAndCashEquivalentsAtCarryingValue', 'PropertyPlantAndEquipmentNet', 'StockholdersEquity', 'AccumulatedDepreciationDepletionAndAmortizationPropertyPlantAndEquipment', 'AccumulatedOtherComprehensiveIncomeLossNetOfTax', 'PropertyPlantAndEquipmentGross'}]
df_concat = pd.concat([df_shareholders.count(), df_shareholders.mean(), df_shareholders.quantile(.25), df_shareholders.quantile(.5), df_shareholders.quantile(.75)], axis=1, keys=['Frenquecy', 'Mean', "Q1", "Median", 'Q3'])
np.round(df_concat, decimals=2)
```
