```python
def get_y(row):
    cik = row['cik']
    corresponding_rows = df[(df['cik'] == cik) & (df['year'] == row['Year'] + 1)]
    # sort rows by month and take the first one
    corresponding_rows = corresponding_rows.sort_values(by=['month'])
    if corresponding_rows.shape[0] == 0:
        return np.nan
    return corresponding_rows.iloc[0]['VALUE']

output_df['eps'] = output_df.apply(get_y, axis=1)

output_df.to_csv("filtered.csv")
```
